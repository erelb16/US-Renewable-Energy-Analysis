 # <b id="section_2">  </b> <b style='color:#06c96b'>3.The Tech Revolution </b> <b></b> 

<div style="color:white;display:fill;border-radius:100px;

```python

legacy_cols = ['Conventional Hydroelectric Power', 'Hydroelectric Power', 'Wood Energy', 'Waste Energy']
tech_cols = ['Solar Energy', 'Wind Energy', 'Geothermal Energy']

df_yearly = df.groupby('Year')[legacy_cols + tech_cols + ['Total Renewable Energy']].sum().reset_index()

df_yearly['Legacy Renewables'] = df_yearly[legacy_cols].sum(axis=1)
df_yearly['Tech Renewables'] = df_yearly[tech_cols].sum(axis=1)

df_yearly['Legacy Share'] = df_yearly['Legacy Renewables'] / (df_yearly['Legacy Renewables'] + df_yearly['Tech Renewables'])
df_yearly['Tech Share'] = df_yearly['Tech Renewables'] / (df_yearly['Legacy Renewables'] + df_yearly['Tech Renewables'])

last_year = 2023
start_year = 2013
years_diff = last_year - start_year

val_start_tech = df_yearly.loc[df_yearly['Year'] == start_year, 'Tech Renewables'].values[0]
val_end_tech = df_yearly.loc[df_yearly['Year'] == last_year, 'Tech Renewables'].values[0]
cagr_tech = (val_end_tech / val_start_tech) ** (1/years_diff) - 1

val_start_legacy = df_yearly.loc[df_yearly['Year'] == start_year, 'Legacy Renewables'].values[0]
val_end_legacy = df_yearly.loc[df_yearly['Year'] == last_year, 'Legacy Renewables'].values[0]
cagr_legacy = (val_end_legacy / val_start_legacy) ** (1/years_diff) - 1

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(16, 20)) 

ax1.plot(df_yearly['Year'], df_yearly['Legacy Renewables'], color='#7f7f7f', label='Legacy (Hydro, Wood, Waste)', linewidth=3, alpha=0.8)
ax1.plot(df_yearly['Year'], df_yearly['Tech Renewables'], color='#2ca02c', label='Tech (Solar, Wind, Geo)', linewidth=4)
ax1.set_title('1. The Divergence: Stagnant Legacy vs. Booming Tech (1973-2023)', fontsize=20, pad=20)
ax1.set_ylabel('Energy Consumption (Trillion Btu)', fontsize=14)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.legend(fontsize=14)

ax1.fill_between(df_yearly['Year'], df_yearly['Tech Renewables'], df_yearly['Legacy Renewables'], 
                 where=(df_yearly['Tech Renewables'] > df_yearly['Legacy Renewables']), 
                 color='#2ca02c', alpha=0.1, interpolate=True)
ax1.text(2015, 500, 'The "Tech" Takeover', color='#2ca02c', fontsize=16, fontweight='bold')


ax2.stackplot(df_yearly['Year'], df_yearly['Legacy Share'], df_yearly['Tech Share'],
              labels=['Legacy Market Share', 'Tech Market Share'], colors=['#e0e0e0', '#98df8a'], alpha=0.9)
ax2.set_title('2. Market Share Evolution: The Shift in Dominance', fontsize=20, pad=20)
ax2.set_ylabel('Share of Total Renewable Energy', fontsize=14)
ax2.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
ax2.legend(loc='lower left', fontsize=12)
ax2.set_xlim(df_yearly['Year'].min(), df_yearly['Year'].max())


bars = ax3.bar(['Legacy Renewables', 'Tech Renewables'], [cagr_legacy, cagr_tech], 
               color=['#7f7f7f', '#2ca02c'], alpha=0.8, width=0.6)
ax3.set_title(f'3. The "Why": 10-Year Growth Rate (CAGR 2013-2023)', fontsize=20, pad=20)
ax3.set_ylabel('Average Annual Growth', fontsize=14)
ax3.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))
ax3.axhline(0, color='black', linewidth=1)

for bar in bars:
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1%}', ha='center', va='bottom' if height > 0 else 'top',
             fontsize=18, fontweight='bold', color='black')

plt.tight_layout(pad=4.0) 
plt.show()
```

### Insight 1: The Technological Revolution â€“ Scalability Wins

The data reveals a decisive shift in the US energy landscape: a transition from resource-constrained 'Legacy' renewables (Hydro, Wood) to scalable 'Tech' renewables (Solar, Wind). While Legacy sources have reached saturation due to physical limitations (e.g., lack of new sites for dams), Tech sources are driving exponential growth. The analysis confirms a 'Great Overtake': Tech renewables have not only surpassed Legacy sources in total production but are now the sole engine of growth, boasting an 11.7% CAGR over the last decade, compared to a -1.8% decline in traditional sectors.

```python
# ---------------------------------------------------------
# Insight 2: The Sector Gap (Residential vs. Industrial) - FIXED
# ---------------------------------------------------------

# 1. Data Prep
# ---------------------------------------------------
target_sectors = ['Residential', 'Industrial']
df_sectors = df[df['Sector'].isin(target_sectors)].copy()
df_sec_yearly = df_sectors.groupby(['Year', 'Sector'])[['Solar Energy', 'Wood Energy', 'Biomass Energy', 'Total Renewable Energy']].sum().reset_index()

# Prepare 2023 Snapshot
df_2023 = df_sectors[df_sectors['Year'] == 2023].groupby('Sector').sum(numeric_only=True).reset_index()

def get_mix(sector_name):
    cols_to_check = ['Hydroelectric Power', 'Geothermal Energy', 'Solar Energy', 'Wind Energy', 'Wood Energy', 'Waste Energy']
    row = df_2023[df_2023['Sector'] == sector_name].iloc[0]
    mix = row[cols_to_check].sort_values(ascending=False)
    return mix[mix > 0] # Keep only non-zero

res_mix = get_mix('Residential')
ind_mix = get_mix('Industrial')

# ×¤×•× ×§×¦×™×” ×©×ž×¦×™×’×” ××—×•×–×™× ×¨×§ ×× ×”× ×ž×¢×œ ×¡×£ ×ž×¡×•×™× (×ž×•× ×¢ ×‘×œ××’×Ÿ)
def my_autopct(pct):
    return ('%1.1f%%' % pct) if pct > 3 else ''

# 2. Visualization
# -------------------------------------------------
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(16, 24))

# --- Chart 1: Line Chart ---
res_data = df_sec_yearly[df_sec_yearly['Sector'] == 'Residential']
ind_data = df_sec_yearly[df_sec_yearly['Sector'] == 'Industrial']

ax1.plot(res_data['Year'], res_data['Solar Energy'], color='#fdbf11', label='Residential Solar', linewidth=5)
ax1.plot(ind_data['Year'], ind_data['Solar Energy'], color='#6b4c35', label='Industrial Solar', linewidth=3, linestyle='--')
ax1.set_title('1. The Adoption Gap: Who is buying Solar?', fontsize=22, pad=20, fontweight='bold')
ax1.set_ylabel('Solar Energy (Trillion Btu)', fontsize=16)
ax1.legend(fontsize=16, loc='upper left')
ax1.grid(True, alpha=0.3)
ax1.text(2018, 150, 'Exponential Growth\n(Households)', color='#e6a800', fontweight='bold', fontsize=16, ha='center')

# --- Chart 2: Residential (Donut) ---
colors_map = {'Solar Energy': '#fdbf11', 'Wood Energy': '#8c564b', 'Geothermal Energy': '#d62728', 'Wind Energy': '#1f77b4', 'Hydroelectric Power': '#aec7e8', 'Waste Energy': 'grey'}
res_colors = [colors_map.get(idx, 'lightgrey') for idx in res_mix.index]

wedges, texts, autotexts = ax2.pie(res_mix, labels=None, autopct=my_autopct, startangle=140, colors=res_colors, pctdistance=0.85, textprops={'fontsize': 14, 'weight': 'bold'})
ax2.set_title('2. Residential Mix (2023): Diverse & High-Tech', fontsize=22, fontweight='bold', pad=20)
ax2.legend(wedges, res_mix.index, title="Energy Sources", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12)
ax2.add_artist(plt.Circle((0,0),0.60,fc='white'))
ax2.text(0, 0, "Solar\nLeader", ha='center', va='center', fontsize=16, fontweight='bold', color='#fdbf11')

# --- Chart 3: Industrial (Donut) - FIXED ---
ind_colors = [colors_map.get(idx, 'lightgrey') for idx in ind_mix.index]

# ×©×™× ×•×™: Labels=None ×›×“×™ ×œ×ž× ×•×¢ ×—×¤×™×¤×”, ×•×”×¢×‘×¨×ª ×”×ž×§×¨× ×œ×¦×“
wedges, texts, autotexts = ax3.pie(ind_mix, labels=None, autopct=my_autopct, startangle=140, colors=ind_colors, pctdistance=0.80, textprops={'fontsize': 14, 'weight': 'bold'})

ax3.set_title('3. Industrial Mix (2023): Dominated by Old Biomass', fontsize=22, fontweight='bold', pad=20)
# ×”×•×¡×¤×ª ×ž×§×¨× ×‘×¦×“ ×‘×ž×§×•× ×›×™×ª×•×‘×™× ×¢×œ ×”×’×¨×£
ax3.legend(wedges, ind_mix.index, title="Energy Sources", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=12)

ax3.add_artist(plt.Circle((0,0),0.60,fc='white'))
ax3.text(0, 0, "Biomass\nMonopoly", ha='center', va='center', fontsize=16, fontweight='bold', color='#6b4c35')

plt.tight_layout(pad=5.0)
plt.show()
```

### Insight 2: The Adoption Gap â€“ Homes Go High-Tech, Industry Stays Traditional
The sectoral analysis reveals a dramatic technological adoption gap. While the Residential sector acts as the growth engine for the solar revolution, boasting exponential adoption and a >30% share in its energy mix, the Industrial sector remains frozen in time, relying almost exclusively (~87%) on traditional biomass. The data proves that the green transition is uneven: modern technologies have successfully penetrated the consumer market but have so far failed to overcome the structural barriers of heavy industry.

A common worry is that renewable energy is 'unreliable' because the sun doesn't always shine. Our analysis puts this fear to rest by revealing a perfect 'Natural Battery' effect. It turns out that nature balances itself: when solar power drops in the winter, wind and water power rise to fill the gap. The data proves that by combining these sources, we get a 'teamwork' effect that makes the power grid 60% more stable and reliable. This means we can maintain a steady energy supply without relying heavily on expensive, man-made battery storage.

This project presents a deep-dive data analysis of the US renewable energy market over 50 years (1973-2023), revealing a definitive "Technological Revolution": a shift from resource-constrained legacy sources (like hydro and wood) to scalable technologies (solar and wind), which have demonstrated exponential growth of ~11.7% annually over the last decade. However, the sectoral breakdown exposes an uneven adoption landscape; while the Residential sector acts as the engine of growth with rapid solar adoption (>30% of its mix), the Industrial sector remains "frozen in time," relying almost exclusively (87%) on traditional biomass due to infrastructure barriers.

Furthermore, the analysis addresses the challenge of grid intermittency by uncovering the "Natural Battery" effect. The data demonstrates a perfect seasonal complementarity between energy sourcesâ€”solar production peaks in the summer precisely when wind and hydro outputs dip, and vice versa. This combination reduces grid volatility by approximately 60%, statistically proving that a diversified renewable mix creates a stable and reliable power supply without heavy reliance on expensive external storage solutions.
